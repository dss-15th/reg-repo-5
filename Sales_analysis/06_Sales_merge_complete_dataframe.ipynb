```python
movie = pd.read_csv('./datas/final_movie_Rev3.csv', index_col=[0])
movie.reset_index(inplace=True)
```

```python
movie.drop(index=0, inplace=True)
movie.rename(columns={'전국': 'Screen', '전국.1': 'Sales', '전국.2': 'Audience', '영화명': 'Movie', '감독': 'Producer', '배급사': 'Distributor', '등급': 'Rate', '장르': 'Genre', '개봉일': 'Open'}, inplace=True)
```

```python
movie['Screen'] = movie['Screen'].apply(lambda x: x.replace(",", "")).astype('int')
movie['Sales'] = movie['Sales'].apply(lambda x: x.replace(",", "")).astype('int64')
movie['Audience'] = movie['Audience'].apply(lambda x: x.replace(",", "")).astype('int')
```

```python
movie['Open'] = pd.to_datetime(movie['Open'])
```

```python
movie['Year'] = pd.DatetimeIndex(movie['Open']).year
movie['Month'] = pd.DatetimeIndex(movie['Open']).month
movie.drop(columns=['Open'], inplace=True)
movie['Season'] = [(month%12 + 3)//3 for month in movie['Month']]
```

```python
movie.reset_index(drop=True, inplace=True)
```

```python
movie['Audience'] = movie['Audience'] / 1000
```

```python
movie.to_csv('./datas/final_movie_Rev4.csv')
```

```python
actors = pd.read_csv('./datas/movie_actors_Rev3.csv', index_col=[0])
actors.rename(columns={'영화명': 'Movie'}, inplace=True)
actors
```

```python
movie = pd.merge(movie, actors, on='Movie')
movie
```

```python
CJEnM = movie[movie['Distributor'] == '씨제이이앤엠(주)']
CJEnM['CJEnM'] = 1
LotteShop = movie[movie['Distributor'] == '롯데쇼핑㈜롯데엔터테인먼트']
LotteShop['LotteShop'] = 1
Next = movie[movie['Distributor'] == '(주)넥스트엔터테인먼트월드(NEW)']
Next['Next'] = 1
ShowBox = movie[movie['Distributor'] == '(주)쇼박스']
ShowBox['ShowBox'] = 1
CJEnt = movie[movie['Distributor'] == '씨제이엔터테인먼트']
CJEnt['CJEnt'] = 1
CJEnM = CJEnM[['Movie', 'CJEnM']]
LotteShop = LotteShop[['Movie', 'LotteShop']]
Next = Next[['Movie', 'Next']]
ShowBox = ShowBox[['Movie', 'ShowBox']]
CJEnt = CJEnt[['Movie', 'CJEnt']] 
movie = pd.merge(movie, CJEnM, on="Movie", how='left') # 58 columns
movie = pd.merge(movie, LotteShop, on="Movie", how='left')
movie = pd.merge(movie, Next, on="Movie", how='left')
movie = pd.merge(movie, ShowBox, on="Movie", how='left')
movie = pd.merge(movie, CJEnt, on="Movie", how='left')
movie = movie.fillna(0)
```

```python
movie
```

```python
convert = {'CJEnM': int, 'LotteShop': int, 'Next': int, 'ShowBox': int, 'CJEnt': int}
movie = movie.astype(convert)
```
